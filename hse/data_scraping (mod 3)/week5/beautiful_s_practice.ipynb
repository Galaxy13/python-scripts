```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Rust_(programming_language)"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text)
```

```python
soup.attrs
```

```python
soup.find_all(name=['div', 'p'], attrs={'href': "/wiki/Category:Pattern_matching_programming_languages"})
```

```python
import re
```

```python
results = soup.select('div[id=bodyContent] p')
for x in results[:10]:
    print(x.text)
```

```python
def filter_function(tag):
    classes = tag.get('class')
    if not classes:
        return False
    return len(classes) > 8
```

```python
soup.find_all(filter_function)
```

```python
links = soup.find_all('img', attrs={'src': True})
```

```python
for link in links:
    url = 'https:' + link['src']
    name = url.split('/')[-1]
    try:
        response = requests.get(url)
        with open(name, 'wb') as f:
            f.write(response.content)
    except:
        print(f'Did not succeed with {url}')
```
