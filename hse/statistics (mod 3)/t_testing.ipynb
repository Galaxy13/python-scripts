```python
import numpy as np
```

```python
import scipy.stats
```

```python
sample_size = 3
exp_uniform = 0.5
iterations = 10000
```

```python
tvalues = []
for i in range(iterations):
    sample = np.random.uniform(0, 1, sample_size)
    t = ((np.mean(sample) - exp_uniform) / np.sqrt(np.var(sample, ddof=1)) * np.sqrt(sample_size))
    tvalues.append(t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(tvalues, bins=500, density=True)
x = np.linspace(-7, 7, 200)
plt.plot(x, scipy.stats.norm.pdf(x))
plt.xlim(-7, 7)
```

```python
plt.hist(tvalues, bins=500, density=True)
x = np.linspace(-7, 7, 200)
plt.plot(x, scipy.stats.t.pdf(x, df=sample_size-1))
plt.xlim(-7, 7)
```

```python
plt.plot(x, scipy.stats.t.pdf(x, df=1), label='df1')
plt.plot(x, scipy.stats.t.pdf(x, df=5), label='df5')
plt.plot(x, scipy.stats.t.pdf(x, df=20), label='df20')
plt.plot(x, scipy.stats.norm.pdf(x), label='norm')
plt.legend()
```
