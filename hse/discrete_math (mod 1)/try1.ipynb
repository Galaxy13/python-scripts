```python
import networkx as nx
```

```python
myGraph = nx.Graph()

myGraph.add_edge('A', 'B')
myGraph.add_edge('B', 'C')
myGraph.add_edge('C', 'D')
myGraph.add_edge('D', 'A')
```

```python
myDirectedGraph = nx.DiGraph() # directed graph
```

```python
myGraph.add_edge('C', 'A', weight=6) # weighted edge (like time of road)
```

```python
nx.read_edgelist # read graph from file
nx.draw_networkx(myGraph)
```

```python
av_clust = nx.average_clustering(myGraph) # average value of local clustering coefficient
av_diam = nx.diameter(myGraph) # diameter of graph
nx.eccentricity # eccentricity dictionary
nx.shortest_path_length # shortest path between vertices
nx.has_path # is there a path
nx.shortest_path # returns shortest path in graph
nx.dfs_preorder_nodes # yields vertices with dfs principle
```
